```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow import keras
from keras.models import Model
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.model_selection import train_test_split
from keras.layers import Dense, Dropout, Conv1D, MaxPooling1D, Flatten,Input, concatenate
from keras.optimizers import Adam
from keras.callbacks import EarlyStopping, ModelCheckpoint
from keras.regularizers import l2

# Load the dataset
data = pd.read_csv('/content/winequality-white-balanced.csv')
data.head()
# Summary statistics
data.describe()
# Separate 'quality' as the target variable and the rest as features
y = data.quality # set 'quality' as target
X = data.drop('quality', axis=1) # rest are features
print(y.shape, X.shape)
# Correlation matrix
colormap = plt.cm.viridis
plt.figure(figsize=(12, 12))
plt.title('Correlation of Features', y=1.05, size=15)
sns.heatmap(data.astype(float).corr(), linewidths=0.1, vmax=1.0,square=True,linecolor='white', annot=True)

# Train-test split
seed = 8 # set seed for reproducibility
X_train, X_test, y_train, y_test = train_test_split(X, y,
test_size=0.2, random_state=seed)
# Scaling the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
# One-hot encode the target for neural networks
encoder = OneHotEncoder()

y_train_encoded = encoder.fit_transform(y_train.values.reshape(-1,1)).toarray()
y_test_encoded = encoder.transform(y_test.values.reshape(-1,1)).toarray()
# Reshape the input for CNN
X_train_scaled_cnn = X_train_scaled.reshape(X_train_scaled.shape[0],X_train_scaled.shape[1], 1)
X_test_scaled_cnn = X_test_scaled.reshape(X_test_scaled.shape[0],X_test_scaled.shape[1], 1)

# Define the CNN layers
def create_cnn_branch(input_layer):
  conv1 = Conv1D(32, kernel_size=2, activation='relu',
  padding='same', kernel_regularizer=l2(0.01))(input_layer)
  maxpool1 = MaxPooling1D(pool_size=2)(conv1)
  conv2 = Conv1D(64, kernel_size=2, activation='relu',
  padding='same', kernel_regularizer=l2(0.01))(maxpool1)
  maxpool2 = MaxPooling1D(pool_size=2)(conv2)
  flatten = Flatten()(maxpool2)
  return flatten
# Split the features into two channels
num_features_channel1 = X_train_scaled_cnn.shape[1] // 2
channel1_input = Input(shape=(num_features_channel1, 1))
channel2_input = Input(shape=(X_train_scaled_cnn.shape[1] -
num_features_channel1, 1))
# Create CNN branches
flatten_1_channel1 = create_cnn_branch(channel1_input)
flatten_1_channel2 = create_cnn_branch(channel2_input)
# Concatenate the output of both channels
merged_cnn = concatenate([flatten_1_channel1, flatten_1_channel2])
# Dense layers for CNN
dense_1_cnn = Dense(256, activation='relu',kernel_regularizer=l2(0.01))(merged_cnn)
dropout_1_cnn = Dropout(0.5)(dense_1_cnn)
dense_2_cnn = Dense(128, activation='relu',kernel_regularizer=l2(0.01))(dropout_1_cnn)
dropout_2_cnn = Dropout(0.5)(dense_2_cnn)
dense_3_cnn = Dense(64, activation='relu',kernel_regularizer=l2(0.01))(dropout_2_cnn)
dropout_3_cnn = Dropout(0.5)(dense_3_cnn)
# Output layer for CNN

output_cnn = Dense(y_train_encoded.shape[1],activation='softmax')(dropout_3_cnn)
# Create the CNN model
cnn_model = Model(inputs=[channel1_input, channel2_input],outputs=output_cnn)
# Compile the CNN model
cnn_model.compile(optimizer=Adam(learning_rate=0.0005),loss='categorical_crossentropy', metrics=['accuracy'])
# Train the CNN model
early_stopping = EarlyStopping(monitor='val_loss', patience=10,restore_best_weights=True)
model_checkpoint = ModelCheckpoint('best_cnn_model.h5',save_best_only=True, monitor='val_loss')
history_cnn = cnn_model.fit(
[X_train_scaled_cnn[:, :num_features_channel1, :],
X_train_scaled_cnn[:, num_features_channel1:, :]],
y_train_encoded,
epochs=10,
batch_size=64,
validation_split=0.2,
callbacks=[early_stopping, model_checkpoint],
verbose=1
)
# Load the best CNN model
cnn_model.load_weights('best_cnn_model.h5')
# Evaluate the CNN model
cnn_scores = cnn_model.evaluate(
[X_test_scaled_cnn[:, :num_features_channel1, :],
X_test_scaled_cnn[:, num_features_channel1:, :]],
y_test_encoded,
verbose=1
)
print("CNN Model Testing Accuracy:", cnn_scores[1])
# MLP model
from sklearn.neural_network import MLPClassifier
# Instantiate the MLP model
mlp_model = MLPClassifier(hidden_layer_sizes=(256, 128, 64),
max_iter=1000, random_state=seed)
# Train MLP model

mlp_model.fit(X_train_scaled, y_train)
# Extract MLP features
mlp_features_train = mlp_model.predict_proba(X_train_scaled)
mlp_features_test = mlp_model.predict_proba(X_test_scaled)
# Evaluate the MLP model
mlp_testing_accuracy = mlp_model.score(X_test_scaled, y_test)
print("MLP Model Testing Accuracy:", mlp_testing_accuracy)
# Create input for the combined model
mlp_output = Input(shape=(mlp_features_train.shape[1],))
# Concatenate the output of CNN and MLP
cnn_flatten_output = Flatten()(output_cnn)
merged_output = concatenate([cnn_flatten_output, mlp_output])
# Dense layers for combined model
dense_1 = Dense(128, activation='relu',
kernel_regularizer=l2(0.01))(merged_output)
dropout_1 = Dropout(0.5)(dense_1)
dense_2 = Dense(64, activation='relu',
kernel_regularizer=l2(0.01))(dropout_1)
dropout_2 = Dropout(0.5)(dense_2)
dense_3 = Dense(32, activation='relu',
kernel_regularizer=l2(0.01))(dropout_2)
dropout_3 = Dropout(0.5)(dense_3)
# Output layer for combined model
output = Dense(y_train_encoded.shape[1],
activation='softmax')(dropout_3)
# Create the combined model
combined_model = Model(inputs=[channel1_input, channel2_input,
mlp_output], outputs=output)
# Compile the combined model
combined_model.compile(optimizer=Adam(learning_rate=0.0005),
loss='categorical_crossentropy', metrics=['accuracy'])
# Train the combined model
history_combined = combined_model.fit(
[X_train_scaled_cnn[:, :num_features_channel1, :],
X_train_scaled_cnn[:, num_features_channel1:, :], mlp_features_train],
y_train_encoded,
epochs=50,
batch_size=64,
validation_split=0.2,

callbacks=[early_stopping, model_checkpoint],
verbose=1
)
# Evaluate the combined model
scores_combined = combined_model.evaluate(
[X_test_scaled_cnn[:, :num_features_channel1, :],
X_test_scaled_cnn[:, num_features_channel1:, :], mlp_features_test],
y_test_encoded,
verbose=1
)
print("Combined Model Testing Accuracy:", scores_combined[1])
```
